# Scraping Huizenzoeker.nl to Analyse the Dutch Housing Market

Here we import both datasets in the form of CSV files from RStudio after they have been modified to work like real datasets (e.g. adjusting variable classes from character to numeric, and fixing the thousands/decimals delimiters). 

```python
#Import packages
import pandas as pd 
```

**Municipality-level dataset**

```python
huizenzoeker_municipality = pd.read_csv('../../data/huizenzoeker_municipality.csv', encoding= 'latin-1') 
```

```python
huizenzoeker_municipality = pd.DataFrame(huizenzoeker_municipality)
huizenzoeker_municipality
```

```python
huizenzoeker_municipality.describe() #to return the summary statisitics for all municipalities in the Netherlands. 
```

**Province-level dataset**

```python
huizenzoeker_province = pd.read_csv('../../data/huizenzoeker_province.csv', encoding= 'latin-1')
```
